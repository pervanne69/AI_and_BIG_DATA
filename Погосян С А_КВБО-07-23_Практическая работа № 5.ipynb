```python
import math

class Trigonometry:
    @staticmethod
    def cos(angle_rad: float) -> float:
        """Вычисление косинуса угла в радианах."""
        return math.cos(angle_rad)
    
    @staticmethod
    def sin(angle_rad: float) -> float:
        """Вычисление синуса угла в радианах."""
        return math.sin(angle_rad)
    
    @staticmethod
    def tan(angle_rad: float) -> float:
        """Вычисление тангенса угла в радианах."""
        return math.tan(angle_rad)
    
    @staticmethod
    def arcsin(value: float) -> float:
        """Вычисление арксинуса значения (возвращает угол в радианах)."""
        if -1 <= value <= 1:
            return math.asin(value)
        else:
            raise ValueError("Значение должно быть в диапазоне от -1 до 1.")
    
    @staticmethod
    def arccos(value: float) -> float:
        """Вычисление арккосинуса значения (возвращает угол в радианах)."""
        if -1 <= value <= 1:
            return math.acos(value)
        else:
            raise ValueError("Значение должно быть в диапазоне от -1 до 1.")
    
    @staticmethod
    def arctan(value: float) -> float:
        """Вычисление арктангенса значения (возвращает угол в радианах)."""
        return math.atan(value)
    
    @staticmethod
    def degrees_to_radians(degrees: float) -> float:
        """Перевод угла из градусов в радианы."""
        return math.radians(degrees)

# Пример использования:
angle_in_degrees = 45
angle_in_radians = Trigonometry.degrees_to_radians(angle_in_degrees)

print("Косинус:", Trigonometry.cos(angle_in_radians))
print("Синус:", Trigonometry.sin(angle_in_radians))
print("Тангенс:", Trigonometry.tan(angle_in_radians))
```

```python
import random

class Tree:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
        
    def insert(self, value):
        if value < self.data:
            if self.left is None:
                self.left = Tree(value)
            else:
                self.left.insert(value)
        elif value > self.data:
            if self.right is None:
                self.right = Tree(value)
            else:
                self.right.insert(value)
        else:
            pass
    def print_tree(self, level=0, label="Root"):
        print("    " * level + f"{label}: {self.data}")
        if self.right:
            self.right.print_tree(level + 1, "Right")
        if self.left:
            self.left.print_tree(level + 1, "Left")
            
            
tree = Tree(10)
for i in range(20):
    tree.insert(random.randint(1, 100))
tree.print_tree()
```

```python
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
x = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
target = [0, 0, 0, 1, 1, 1]

x_train, x_test, y_train, y_test = train_test_split(x, target, test_size=0.2)

print(x_train, x_test, y_train, y_test, sep="\n")

```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(x_train, y_train)
```

```python
from sklearn import tree
tree.plot_tree(classifier)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor



dataframe = pd.read_csv("petrol_consumption.csv")
dataframe = dataframe[['Petrol_Consumption', 'Petrol_tax', 
'Average_income']]
print("Dataframe first 5 lines:", dataframe.head(5), sep="\n")
print("Dataframe description: ", dataframe.describe(), sep="\n")
print("Dataframe shape: ", dataframe.shape)


x = dataframe[["Average_income", "Petrol_tax"]]
y = dataframe["Petrol_Consumption"]

```

```python

 
x_train, x_test, y_train, y_test = train_test_split(x, y, 
test_size=0.2, random_state=0)
print("Набор стажа для обучения: ", x_train, 
  "Набор стажа для тестирования:", x_test, 
  "Набор зарплат для обучения:", y_train, 
  "Набор зарплат для тестирования:", y_test, sep="\n")
```

```python
regressor = DecisionTreeRegressor()
regressor.fit(x_train, y_train)
```

```python
from sklearn import tree
tree.plot_tree(regressor)
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
df = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
df
```

```python
from sklearn import metrics
import numpy as np
print("Mean Squared Error: ", metrics.mean_squared_error(y_test, y_pred))
print("Mean Absolute Error: ", metrics.mean_absolute_error(y_test, y_pred))
```
